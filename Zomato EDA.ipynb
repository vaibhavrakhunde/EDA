```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observation

1. When rating is between 4.5 to 4.9 = Excellent Food 
2. when rating is between 4.0 to 4.4 = Very Good 
3. when rating is between 3.5 to 3.9 = Good
4. when rating is between 3.0 to 3.4 = Avg
5. when rating is between 2.5 to 2.9 = Avg
6. when rating is between 2.0 to 2.4 = Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data= ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating count',hue= 'Rating color', data= ratings, palette=['purple','red','orange', 'yellow','green','green'])
```

```python
## Count plot

sns.countplot(x='Rating color', data= ratings, palette=['purple','red','orange', 'yellow','green','green'])
```
